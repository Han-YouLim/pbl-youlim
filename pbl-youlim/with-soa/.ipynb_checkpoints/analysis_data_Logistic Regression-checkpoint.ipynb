# 1.데이터 읽기

```python
import warnings
warnings.filterwarnings(action='ignore')
import pandas as pd
```

```python
def kaggle_format(df):
    df['label'][df['label'] == 'none'] = 0
    df['label'][df['label'] == 'offensive'] = 1
    return df
```

```python
df = pd.read_csv('./data/1cha_dataset.csv',sep = ',',
                 names=['comments', 'label'], header=None, quoting=3)
df.head()
```

```python
#특수문자 제거
import re

df_only_comments = df.iloc[:,[0]]
print(df_only_comments)
comment_without_Special_Characters  = []
for comment in df['comments'][:]:
    comment_without_Special_Characters.append(re.sub("[^가-힣 ㄱ-ㅎ ㅏ-ㅣ \\s]", "", comment )) #한글 아닌 것들 다 제거

index = 0
for comment in comment_without_Special_Characters:
    df['comments'][index] = comment
    index += 1

df
```

## 1-1. 데이터 분할

```python
X = df['comments']
y = df['label']
```

## 1-2. 벡터화
 - TFIDF
 -CountVectorizer -> CV

```python
from sklearn.feature_extraction.text import CountVectorizer
vec_cv = CountVectorizer(analyzer = "word", max_features = 6000) 
X_cv = vec_cv.fit_transform(X)
```

```python
X_cv.shape
```

```python
#특징 추출
from sklearn.feature_extraction.text import TfidfVectorizer

vec_tf = TfidfVectorizer(min_df = 0.0, analyzer = 'char', sublinear_tf=True, 
    ngram_range = (1,3), max_features=5000)


X_tf = vec_tf.fit_transform(X)
```
